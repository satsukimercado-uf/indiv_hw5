Given function f(x1, x2, x3, x4) = (x1*x2+x3)^2/(exp(x4)) + sin(x1+x4)

Rewritten linear function f(c,d,f) = c/d + f

    a = x1*x2 
    b = a + x3
    c = b^2
    d = exp(x4)
    e = x1+x4
    f = sin(e)


```python
# Define input tensors with requires_grad = True
x1 = torch.tensor([1.0], requires_grad= True)
x2 = torch.tensor([2.0], requires_grad= True)
x3 = torch.tensor([3.0], requires_grad= True)
x4 = torch.tensor([4.0], requires_grad= True)
```

```python
# Build the computation graph step by step, defining tensor to decompose f into intermediate tensors 
def computation_graph(x1,x2,x3,x4):
    a = x1*x2
    b = a+x3
    c = torch.square(b)
    d = torch.exp(x4)
    e = x1+x4
    f = torch.sin(e)
    g = torch.divide(c,d)
    f = torch.add(f,g)
    f.requires_grad_()
    return f


# Set randomized initialized values for weight vector 
def init_w():
    torch.manual_seed(1)
    w = torch.empty(4,1)
    nn.init.xavier_normal_(w)
    w.requires_grad_()
    return w

# Set randomized initialized values for bias term
def init_b():
    torch.manual_seed(1)
    b = torch.empty(4,1)
    nn.init.xavier_normal_(b)
    b.requires_grad_()
    return b 

# Compute the gradients of f with respectt to each of the input variables using back propagation, calling .backward()   
def gradients(f,x1,x2,x3,x4):
    f.backward()
    print('df/dx1 =', x1.grad)
    print('df/dx2 =', x2.grad)
    print('df/dx3 =', x3.grad)
    print('df/dx4 =', x4.grad) 

# 
```

```python
# Homework example values
f = computation_graph(x1, x2, x3, x4)
w = init_w()
b = init_b()
x = torch.tensor([[x1],[x2],[x3],[x4]])
z = torch.add(torch.mul(w,x), b)
print("z", z)
gradients(f,x1,x2,x3,x4)

print(w)
print(b)
```
