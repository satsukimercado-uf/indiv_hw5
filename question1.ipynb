Given function f(x1, x2, x3, x4) = (x1*x2+x3)^2/(exp(x4)) + sin(x1+x4)

Rewritten linear function f(c,d,f) = c/d + f

    a = x1*x2 
    b = a + x3
    c = b^2
    d = exp(x4)
    e = x1+x4
    f = sin(e)


```python
# Define input tensors with requires_grad = True
x1 = torch.tensor([1.0], requires_grad= True)
x2 = torch.tensor([2.0], requires_grad= True)
x3 = torch.tensor([3.0], requires_grad= True)
x4 = torch.tensor([4.0], requires_grad= True)
```

```python
# Build the computation graph step by step, defining tensor to decompose f into intermediate tensors 
def computation_graph(x1,x2,x3,x4):
    a = x1*x2
    b = a+x3
    c = torch.square(b)
    d = torch.exp(x4)
    e = x1+x4
    f = torch.sin(e)
    g = torch.divide(c,d)
    z = torch.add(f,g)
    z.requires_grad_()
    return z, a, b, c, d, e, f, g


# Set randomized initialized values for weight vector 
def init_w():
    torch.manual_seed(1)
    w = torch.empty(4,1)
    nn.init.xavier_normal_(w)
    w.requires_grad_()
    return w

# Set randomized initialized values for bias term
def init_b():
    torch.manual_seed(1)
    b = torch.empty(4,1)
    nn.init.xavier_normal_(b)
    b.requires_grad_()
    return b 

# Compute the gradients of f with respect to each of the input variables using back propagation, calling .backward()   
def gradients(z,x1,x2,x3,x4):
    # Call backward
    z.backward()
    print('df/dx1 =', x1.grad)
    print('df/dx2 =', x2.grad)
    print('df/dx3 =', x3.grad)
    print('df/dx4 =', x4.grad) 

# 
```

```python
# Homework example values
z, a, b, c, d, e, f, g = computation_graph(x1, x2, x3, x4)
# w = init_w()
# b = init_b()
z.requires_grad_()
print("z =", z)

# Compute gradients after .backward()
gradients(z,x1,x2,x3,x4)

# What happens to the computation graph
# print(x1, x2, x3, x4)
print("a = ", a, "\nb =", b, "\nc =", c, "\nd =", d, "\ne =", e, "\nf =", f, "\ng =", g)
```
